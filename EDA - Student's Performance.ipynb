## now lets have a grading system for the given dataset

```python
def gradeassign(marks):
    if marks >=90 and marks<=100:
        return 'Grade A1'
    elif marks >=80 and marks<90:
        return 'Grade A2'
    elif marks >=70 and marks<80:
        return 'Grade B1'
    elif marks >=60 and marks<70:
        return 'Grade B2'
    elif marks >=50 and marks<60:
        return 'Grade C1'
    elif marks >=40 and marks<50:
        return 'Grade c2'
    elif marks >= 0 and marks<40:
        return 'Grade D'
```

```python
student['grades']= student['percentage'].apply(gradeassign)
```

```python
student['grades'].value_counts().plot.pie()
```
